##### Using Group-by’s and Merges
Group-by’s can be used to build groups of rows based on a specific feature in your dataset eg. the ‘Manufacturer’ categorical column. 

```python
group_by_Manufacturer = car_data.groupby(by=['Manufacturer'])

car_data_avg = round(group_by_Manufacturer.mean(),0)
car_data_avg
```

```python
# A box plot for the revs per mile for the Audi, Hyundai, Suzuki and Toyota 
# car manufacturers. Which of these manufacturers has the car with the
# highest revs per mile?
manufacturers = ["Audi", "Hyundai", "Suzuki", "Toyota"]
chosen_makes = car_data.loc[car_data["Manufacturer"].isin(manufacturers)]
chosen_makes.boxplot(by= "Manufacturer", column= "Rev.per.mile")
# The single Suzuki has highest revs per mile
```

```python
# A histogram of MPG in the city. On the same axis, show a histogram of
# MPG on the highway. Is it generally more fuel efficient to drive in the
# city or the highway?
data1 = car_data["MPG.city"]
data2 = car_data["MPG.highway"]
plt.figure(figsize=(10,6))
plt.hist(data1, bins=50, alpha=0.5, label="City")
plt.hist(data2, bins=50, alpha=0.5, label="Highway")
plt.legend(loc="upper right")
# Generally highway is more fuel efficient
```

```python
# A lineplot showing the relationship between the Wheelbase and
# turning circle. What is this relationship? What happens when the
# wheelbase gets larger?

# As the wheelbase gets larger so does the turning circle

x= car_data["Turn.circle"]
y = car_data["Wheelbase"]
plt.figure()
sns.lineplot(data = car_data, x="Turn.circle", y="Wheelbase")
```

```python
# A bar plot showing the mean horsepower for each car Type (Small,
# Midsize, etc.). Does a larger car mean more horsepower?
x = car_data["Type"]
y = car_data["Horsepower"].mean
plt.figure()
sns.barplot(data = car_data, x = "Type", y = "Horsepower")
# Larger cars generally have more horsepower
```

```python
# A pie chart for the respective engine sizes. Which engine size seems to
# be the most popular?
e_count = car_data["EngineSize"].value_counts()
plt.figure(figsize= (10,15))
plt.pie(e_count)
plt.pie(e_count, autopct = "%0.1f%%")
plt.legend(car_data["EngineSize"])
plt.show
# 3 Litre is the most common
```
